## 선형그래프


*   기본그래프



```
# 사용 패키지 불러오기
import matplotlib.pyplot as plt
import numpy as np
```

```
data = np.arange(10)
```

```
# 그래프 그리기
plt.title('test')
plt.xlabel('test_x')
plt.ylabel('test_y')
plt.plot(data)
```



*   String Format





```
# 축 범위 지정
plt.axis([0, 4, -2, 64])
# 하나의 차트에 여러 그래프 그리기
plt.plot(data, '<g')
plt.plot(data**2, 'ro-.')
plt.plot(data**3, 'k^--')
```

1.   색
> 그래프 선의 색상이 파랑으로 디폴트 값으로 설정. 지정 문자열을 입력해 변형 가능.

2.   포인트 모양
> 입력 데이터 위치 표시 포인트 모양이 존재하지 않는 것으로 디폴트 값으로 설정. 따라서 그래프 색만 수정 시 연결 선이 사라지지 않으며 포인트 모양 수정 시 연결 선 사라짐.
3.   선
> 디폴트 값은 솔리드라인(기본 그래프 선)으로 설정.

## subplot

```
# figure size = 가로, 세로
plt.figure(figsize=(6,6))

# subplot(n,m,i) : n*m 행렬에서 i번째 셀
plt.subplot(331)
plt.plot(data, '-.')
plt.subplot(335)
plt.plot(data, 'r--')
plt.subplot(339)
plt.plot(data, 'y:')
```

## 그래프 종류

```
names = ['group_a','group_b','group_c']
values = [1, 10, 100]

plt.figure(figsize=(8,8))

# 세로막대
plt.subplot(221)
plt.bar(names, values)

# 가로막대
plt.subplot(222)
plt.barh(names,values)

# 분포그래프
plt.subplot(223)
plt.scatter(names, values)

# 직선그래프
plt.subplot(224)
plt.plot(names, values)

plt.suptitle('Categorical Plotting', color = 'r')
plt.show()
```
