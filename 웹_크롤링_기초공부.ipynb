```
# 웹 크롤링 필요 모듈
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pandas import DataFrame as df
import csv
import datetime
```

```
# 크롤링할 때 목적지 찾는 방법
# 목적지 페이지로 가서 F12 눌러 창에서 숩겨진 url 찾기
```

```
# 코드작성
req = requests.get('https://www.naver.com/')
html = req.content
soup = BeautifulSoup(html, 'html.parser', from_encoding='ecu-kr')
```

```
# body 테그 안의 모든 내용 출력
print(soup.find_all('body'))
```

```
# div 태그 안의 모든 내용 출력
print(soup.find_all('div'))
```
